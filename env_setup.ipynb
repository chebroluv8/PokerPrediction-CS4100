```python
import rlcard
from rlcard.agents import RandomAgent
import numpy as np
```

```python
# Set up environment 
env = rlcard.make('limit-holdem')
agent = RandomAgent(num_actions = env.num_actions)
env.set_agents([agent, agent])   

print("num_actions:", env.num_actions)
print("num_players:", env.num_players)
print("state_shape:", env.state_shape)
print("action_shape:", env.action_shape)
```

```python
# Run one game
trajectories, payoffs = env.run()
print("Trajectories", trajectories)
print("Payoffs:", payoffs)
```

```python
# Run by taking random steps
reward_list = []

for i in range(10000):
    trajectories, payoffs = env.run()
    reward_list.append(payoffs[0])

avg_reward = np.average(reward_list)
print("Average reward obtained:", avg_reward)
```
